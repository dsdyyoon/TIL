```python
# https://teddylee777.github.io/scikit-learn/gradient-descent
    
import numpy as np
import matplotlib.pyplot as plt

x = np.random.rand(100)
y = 0.2 * x + 0.5

plt.figure(figsize = (8,6))
plt.scatter(x, y)
plt.show()
```

```python
# 예측값, 실제값의 함수 
def plot_prediction(pred, y):
    plt.figure(figsize=(8,6))
    plt.scatter(x,y)
    plt.scatter(x, pred)
    plt.show()
```

```python
errors =[]

w = np.random.uniform(-1,1)
b = np.random.uniform(-1,1)

learning_rate=0.7
for epoch in range(100):
    y_pred = w * x + b 

    error = np.abs(y_pred - y).mean()
    if error < 0.001:
        break 
    
    # gd 계산
    w_grad = learning_rate * ((y_pred- y) * x).mean()
    b_grad = learning_rate * (y_pred- y).mean()
    
    errors.append(error)

    # w, b값 갱신
    w = w - w_grad
    b = b - b_grad
    
    # batch_
    if epoch % 5 ==0: #5의 배수마다 확인
        y_pred = w * x + b
        plot_prediction(y_pred, y)
```

```python
plt.figure(figsize=(10, 7))
plt.plot(errors)
plt.xlabel('Epochs')
plt.ylabel('Error')
plt.show()
```
